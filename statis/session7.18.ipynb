```python
movies=pd.read_csv('imdb-top-1000.csv')
movies.sample()
```

```python
genre=movies.groupby('Genre')
```

```python
#Applying builtin aggregation function on groupby objects.
genre.sum()
```

```python
# find the top 3 genre by total earning
movies.groupby('Genre').sum()
```

```python
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False)
```

```python
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
#find the genre with highest avg IMDB rating.
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
#find the genre with highest avg IMDB rating.
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
 #find director with most popularity.
movies.groupby('Director')['Gross'].sum().sort_values(ascending=False)
```
